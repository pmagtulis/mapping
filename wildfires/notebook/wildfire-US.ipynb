# Do your imports

```python
import pandas as pd
import numpy as np
import altair as alt
import geopandas as gpd
pd.set_option('display.max_columns', None)
```

# Read file

We are going to read the Excel file containing data about **wildfire exposure and vulnerability** across US states.

```python
df = pd.read_excel("wildfire.xlsx")
df.head()
```

# Select and isolate a state

The homework said we should select one state and in here we are using the state of **Texas**.

```python
texas = df[df.STATE=='Texas']
texas
```

# Cleaning the data

Now, we are ready to clean the data. After isolating **Texas data** from our main dataframe into a new dataframe, we then proceed with cleaning the dataset that we have for mapping later.

## First stop, get the columns that we need

There are just so many columns in the original dataset and we only need one datapoint for our analysis. We isolate that together with the **GEOID** and **county names**.

```python
texas_fire = texas[['STATE','GEOID', 'NAMELSAD','COUNTYNS','Total number of Housing Units (HUs)', 'Total number of exposed HUs']]
texas_fire
```

## Rename columns for easier analysis

Lowercase all columns and simplify their names for easier coding.

```python
texas_fire.columns = ['state', 'geoid', 'name', 'county_code', 'housing_units', 'exposed_units']
texas_fire
```

## Compute for percentage of houses at risk of burning

We will use the **housing units** data across Texas. We have the absolute number of houses believed to be **directly or indirectly exposed to wildfires.**

But we need the **percentage** instead for better plotting. We therefore compute for that and put the result in a column.

```python
texas_fire['pct_burned'] = texas_fire.exposed_units / texas_fire.housing_units * 100
texas_fire
```

## Round up the 'pct_burned' column

```python
texas_fire.pct_burned = texas_fire.pct_burned.round(1)
texas_fire
```

## Finally, we reset the index

```python
texas_fire.reset_index(inplace=True)
```

```python
texas_fire
```

# Geopandas

We now proceed with our geopandas. First, we read our geometry file containing the coordinates for our maps.

```python
counties = gpd.read_file('https://www2.census.gov/geo/tiger/GENZ2018/shp/cb_2018_us_county_20m.zip')
counties
```

# Merge the datasets

We will need to merge our Texas wildfire dataset with that of our geometrical data for mapping. But first let's check the data type we are dealing with for both.

```python
texas_fire.dtypes
```

```python
counties.dtypes
```

# More cleaning

Two things we need to do: first, lowercase the column names. Second, convert geoid column in **texas_fire** df to string to match that of geopandas df.

```python
counties.columns = counties.columns.str.lower()
counties
```

## Convert GEOID to string

This is to be able to add zeroes to them.

```python
texas_fire.geoid = texas_fire.geoid.astype('str').str.zfill(5)
texas_fire
```

## Actual merging

```python
countiesData = counties.merge(texas_fire, on='geoid')
countiesData
```

# Remove unnecessary columns

We don't need all of them and there are some duplicates.

```python
countiesData = countiesData.drop(['countyfp', 'countyns','statefp', 'affgeoid', 'lsad', 'name_x', 'index'], axis=1)
countiesData
```

# Make your buckets

We will be adding a new column named **percentiles** that will group our other new column **pct_burned** into buckets. This will be used for our mapping.

```python
countiesData.pct_burned.sort_values(ascending=False)
#This means the lowest value is around 25%. Highest is 100%
```

```python
countiesData['percentiles'] = pd.cut(np.array(countiesData['pct_burned']), [0, 11, 21, 31, 41, 51, 61, 71, 81, 91, 100],
                                     labels=["0-10", "11-20", "21-30", "31-40", "41-50", "51-60", "61-70", "71-80", "81-90", "91-100"])

countiesData
```

```python
countiesData.percentiles = countiesData.percentiles.astype('str')
```

## Round off the housing units data

```python
countiesData.housing_units = countiesData.housing_units.round(0)
countiesData.exposed_units = countiesData.exposed_units.round(0)
countiesData
```

# Try it in altair

```python
countiesMap = alt.Chart(countiesData).mark_geoshape().encode(
    color=alt.Color('pct_burned:Q',
      bin=alt.Bin(maxbins=5, step=10),
      legend=alt.Legend(title='Vulnerable housing to wildfire'),
      scale=alt.Scale(scheme='lightorange')),
    tooltip=[alt.Tooltip('name_y',title='County'),
      alt.Tooltip('pct_burned',title='Vulnerable housing to wildfire')]
).project(
    type='albersUsa'
).properties(
    title='More Texas houses at risk from wildfires',
    width=700,
    height=450
).configure_view(
    strokeWidth=0
)
```

```python
countiesMap
```

# Convert to GEOJSON

```python
countiesData.to_file('countiesWildfire.geojson', driver='GeoJSON')
```
