# Read your files

We'll be checking both the shapefile and the CSV which we will combine later.

## Shapefile

```python
countries = gpd.read_file('https://naciscdn.org/naturalearth/110m/cultural/ne_110m_admin_0_countries.zip')
```

```python
countries.head()
```

## CSV 

The dataset contains the **percentage of energy production sourced from oil as of 2015**. Data were collated from the World Bank.

```python
df= pd.read_csv('oil_energy.csv')
df
```

# Cleaning the shapefile

## Remove unnecessary columns

```python
countries= countries.drop(['FCLASS_UA', 'FCLASS_BD', 'FCLASS_SE', 'FCLASS_NL', 'FCLASS_IT', 'FCLASS_GR', 'FCLASS_PL', 'FCLASS_ID'], axis=1)
countries
```

There are so many columns apparently (about 160+) so let's just get the subset that we want to match with our data.

```python
countries_real = countries[["NAME_LONG", "NAME_CIAWF", "INCOME_GRP", "CONTINENT", "geometry"]]
countries_real
```

```python
countries_real.columns = countries_real.columns.str.lower()
countries_real.rename(columns={'name_ciawf':'country'}, inplace=True)
countries_real
```

# Merge with CSV data

```python
df
```

```python
#pd.set_option('display.max_rows', None)
countries_oil = countries_real.merge(df, on='country')
countries_oil
```

# Last checks before mapping

Replace column names, check for info type, look for values relevant for mapping (i.e. which is highest, lowest), deal with NaNs.

## Rename columns

```python
countries_oil.rename(columns={'Power production from oil':'oil_energy'}, inplace=True)
countries_oil
```

## Fill NaN values

```python
countries_oil.oil_energy.isna().sum() #30 countries with no values.
```

```python
countries_oil = countries_oil.fillna(0)
countries_oil
```

## Drop duplicated column

We have two columns with country names. We only need one.

```python
countries_oil = countries_oil.drop(['name_long'], axis=1)
countries_oil
```

## Clean 'income grp' column

Remove the numbers at the start of each entry using regex.

```python
countries_oil.income_grp = countries_oil.income_grp.str.replace(r'^\d[.]', "", regex=True)
countries_oil
```

# Create bins for 'oil_energy' column

This will help us in mapping differences in oil as source of energy per each territory.

```python
countries_oil.oil_energy.sort_values(ascending=False)
```

```python
countries_oil['percentiles'] = pd.cut(np.array(countries_oil['oil_energy']),
   10, labels=["0-10", "11-20", "21-30", "31-40", "41-50", "51-60", "61-70", "71-80", "81-90", "91-100"])

countries_oil
```

```python
countries_oil.percentiles = countries_oil.percentiles.astype(str) #Transform to string so we can convert to geojson.
```

```python
countries_oil.dtypes
```

# Convert to GEOJSON

```python
countries_oil.to_file('countries_oil.geojson', driver='GeoJSON')
```
